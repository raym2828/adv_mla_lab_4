# Setup Steps

1. Create new project folder


```python
mkdir new_project
```

2. Initialise folder with .git

```python
git init
```

3. Login into Github with your account (https://github.com/) and create a public repo with the name
4. Link local repo with the Github. Source Control > Add remote



5. make all files in directory synced

```python
git add .
```

[2.1] Create a file called requirements.txt with the following content:

```python

streamlit==1.36.0
pandas==2.2.2
scikit-learn==1.5.1
```

[2.2] Create a file called Dockerfile with the following content: (Note there is no extention to this file)

```python

FROM python:3.11.4

WORKDIR /app

COPY requirements.txt .

RUN pip install -r requirements.txt

COPY . /app

CMD ["streamlit", "run", "app.py"]
```

[2.3] Create an empty file called app.py

[2.4] Build the image from the Dockerfile

```python
sudo docker buildx build -t streamlit_lab4:latest .
```

[2.5] Run the built image with 8501 port published and a volume mounted to the current directory to /app

```python
docker run -dit --rm --name advmla_lab4_app -p 8501:8501 -v "${PWD}":/app streamlit_lab4:latest
```
